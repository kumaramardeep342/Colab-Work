# 4.You are given a csv file named “EmployeeDetails.csv”. Perform the following tasks:

- Use pandas module to read the csv file
- Split the name as first name and last name and make two separate columns as shown
- Delete “Name” Column
- Increment the “Salary” Column by 10% of the existing valueand save changes in same column onlyas shown
- Save changes to a new file named “Employee_Data.csv”

```python
#enable data table format
from google.colab import data_table
data_table.enable_dataframe_formatter()

#disable datatable format
# from google.colab import data_table
# data_table.disable_dataframe_formatter()

import pandas as pd

#Use pandas module to read the csv file
EmployeeDetails = pd.read_csv('/content/drive/MyDrive/Colab Work/Edureka AI Certification/Dataset/Edureka - M1 - Case Study 1 - EmployeeDetails/EmployeeDetails.csv')
EmployeeDetails.head()
#print(EmployeeDetails.columns)
```

```python
#Split the name as first name and last name and make two separate columns as shown
# in Name column white space is present hence stripping the columns
EmployeeDetails.columns = EmployeeDetails.columns.str.strip()
EmployeeDetails[['First Name', 'Last Name']] = EmployeeDetails['Name'].str.split(' ', expand=True, n=1)
EmployeeDetails = EmployeeDetails[['Name','First Name', 'Last Name', 'Age', 'Salary']].copy()
EmployeeDetails.head()
```

```python
#Delete “Name” Column
EmployeeDetails.drop(columns=['Name'],inplace=True)
EmployeeDetails.head()
```

```python
#Increment the “Salary” Column by 10% of the existing valueand save changes in same column onlyas shown
EmployeeDetails['Salary'] = EmployeeDetails['Salary'] + EmployeeDetails['Salary'] * 0.1
EmployeeDetails['Salary'] = EmployeeDetails['Salary'] .astype('int')
EmployeeDetails.head()
```

```python
EmployeeDetails.to_csv('/content/drive/MyDrive/Colab Work/Edureka AI Certification/Dataset/Edureka - M1 - Case Study 1 - EmployeeDetails/Employee_Data.csv')
```
